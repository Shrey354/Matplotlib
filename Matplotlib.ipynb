```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
y = np.sin(x)
```

```python
z = np.cos(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin wave')
plt.title('sin wave')
plt.show()
```

```python
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y ,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, np.sin(x), 'g+')
plt.plot(x, np.cos(x), 'r.')
plt.show()
```

```python
#Bar plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.bar(langs,students)
plt.xlabel('languages')
plt.ylabel('students')
plt.title('student leaning languages')
plt.show()
```

```python
#Piechart
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [40,17,35,45,12]
ax.pie(students,labels=langs, autopct='%1.1f%%')
plt.show()
```

```python
#Scatterplot
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color='r')
ax.scatter(x,z,color='y')
plt.show()
```

```python
#3D scatter plot
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(50)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c=z, cmap='Blues')
plt.show()
```
